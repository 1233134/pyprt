```python
v = [0, 0, 0,  0, 0, 20,  10, 0, 10,  10, 0, 0]
initialGeometry = pyprt.Geometry(v)
rpk = asset_file("simple_rule2019.rpk")
attrs = ["ruleFile:string=bin/simple_rule2019.cgb", "startRule:string=Default$Footprint"]

mod = pyprt.ModelGenerator([initialGeometry])
generated_model = mod.generate_model(rpk, attrs)

visualize_PRT_results(generated_model)
```

```python
summarize_report(combine_reports(generated_model[0]))
```

#### Test the rule I used for my Deep Learning project:

```python
attrs2 = attrs
attrs2.append("maxHeight:float=9")

mod2 = pyprt.ModelGenerator([initialGeometry, initialGeometry])
generated_model2 = mod2.generate_model(rpk, attrs)

visualize_PRT_results(generated_model2)
```

```python
initialGeometry2 = pyprt.Geometry([0, 0, 0,  0, 0, 10,  10, 0, 10,  10, 0, 0])
initialGeometry3 = pyprt.Geometry([0, 0, 0,  0, 0, 30,  10, 0, 10,  10, 0, 0])

rpk3 = asset_file("simpleRuleDataset.rpk")
attrs3 = ["ruleFile:string=bin/simpleRuleDataset.cgb", "startRule:string=Default$Lot"]
```

```python
reports = []
model_to_generate = pyprt.ModelGenerator([initialGeometry, initialGeometry2, initialGeometry3])
for val in range(0,10):
    attribut_to_vary = "maxHeight:float=" + str(val)
    attrs3.append(attribut_to_vary)
    
    models = model_to_generate.generate_model(rpk3, attrs3)
    
    attrs3.remove(attribut_to_vary)

    if len(models) > 0:
        for m in models:
            reports.append(summarize_report(combine_reports(m)))
```

#### Transform report in pandas dataframe in order to process the dataset before training any ML/DL model on it:

```python
reports_df = pd.DataFrame(reports)
reports_df
```
